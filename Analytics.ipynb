## Wilcoxon Rank Sum Test

```python
from scipy.stats import ranksums
for treatment in range(1,7):
    print("treatment ",treatment)
    print(ranksums(df.buttonpresses[df.treat==treatment], df.buttonpresses[df.treat==7], alternative='greater' ))
```

```python
from scipy.stats import ranksums

index_list = ['T' + str(i) for i in range(1,8)]
wilcoxon_df = pd.DataFrame(index=index_list,columns=index_list)

for treatment in range(1,8):
    for treatment_comparison in range(1,8):
        pval = ranksums(df.buttonpresses[df.treat==treatment], 
                 df.buttonpresses[df.treat==treatment_comparison], 
                 alternative='greater')[1]
        wilcoxon_df.iloc[treatment-1,treatment_comparison-1] = pval


wilcoxon_df
```

The Hypothesis being test is "Row <= Column"

```python
print(wilcoxon_df.to_latex())
```

```python
for treatment1 in range(1,7):
    print("> Treatment ", treatment1)
    for treatment in range(1,7):
        print("treatment ",treatment)
        print(ranksums(df.buttonpresses[df.treat==treatment], df.buttonpresses[df.treat==treatment1], alternative='greater' ))
    print("\n")
```
